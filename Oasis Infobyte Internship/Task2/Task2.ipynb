```python
df.columns=["State", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Area"]
plt.title("Unemployment in India")
sns.histplot(data=df, x="Estimated Employed", hue="Area")
plt.show()
     
```

```python
plt.title("Unemployment in India")
sns.histplot(data=df, x="Estimated Unemployment Rate", hue="Area")
plt.show()
```
